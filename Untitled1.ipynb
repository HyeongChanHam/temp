```python
mu_0, lambda_, alpha, beta = 4, 1.2, 1.3, 1.5


mu = np.linspace(-4, 7, 100)
sigma2 = np.linspace(0., 3, 100)
mu_grid, sigma2_grid = np.meshgrid(mu, sigma2)

pdf_values = np.zeros_like(mu_grid)

for i in range(mu_grid.shape[0]):
    for j in range(mu_grid.shape[1]):
        mu_val = mu_grid[i, j]
        sigma2_val = sigma2_grid[i, j]
        # Normal-Inverse Gamma PDF (up to proportionality constant)
        # pdf_values[i, j] = (sigma2_val ** (-alpha - 1)) * np.exp(-beta / sigma2_val) * \
        #                    norm.pdf(mu_val, mu_0, np.sqrt(sigma2_val / lambda_))
        pdf_values[i, j] = invgamma.pdf(sigma2_val, alpha, scale=beta) * norm.pdf(mu_val, mu_0, np.sqrt(sigma2_val / lambda_))

cp = plt.contourf(mu, sigma2, pdf_values, levels=50, cmap='viridis')
plt.colorbar(cp)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
import ipywidgets as widgets

#   
def normal_distribution(x, mean, std_dev):
    return (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev)**2)

#   
def plot_normal_distribution(mean, variance):
    std_dev = np.sqrt(variance)  #   
    x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)  # x   
    y = normal_distribution(x, mean, std_dev)  # y  

    #  
    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label=f'Normal Distribution\nMean={mean:.2f}, Variance={variance:.2f}')
    plt.title('Interactive Normal Distribution')
    plt.xlabel('X')
    plt.ylabel('Probability Density')
    plt.xlim(-10, 10)
    plt.ylim(0, 1)
    plt.legend()
    plt.grid()
    plt.show()

#     
interact(
    plot_normal_distribution,
    mean=FloatSlider(value=0, min=-10, max=10, step=0.1, description='Mean'),
    variance=FloatSlider(value=1, min=0.1, max=10, step=0.1, description='Variance')
)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

#   
def normal_distribution(x, mean, std_dev):
    return (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev)**2)

# |X|  
def abs_normal_distribution(x, mean, std_dev):
    pdf = normal_distribution(x, mean, std_dev)
    pdf[x < 0] = 0  # x < 0 PDF 0
    return 2 * pdf  # |X| PDF x >= 0 f_X(x)  

#   
def plot_abs_normal_distribution(mean, variance):
    std_dev = np.sqrt(variance)  #   
    x = np.linspace(-4*std_dev, 4*std_dev, 1000)  # x   
    y = normal_distribution(x, mean, std_dev)  # X PDF
    abs_x = np.linspace(0, 4*std_dev, 500)  # |X|  
    abs_y = abs_normal_distribution(abs_x, mean, std_dev)  # |X| PDF

    #  
    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label='Normal Distribution $f_X(x)$')
    plt.plot(abs_x, abs_y, label='|X| Distribution $f_{|X|}(x)$', linestyle='--')
    plt.title('Normal Distribution and |X| Distribution')
    plt.xlabel('X')
    plt.ylabel('Probability Density')
    plt.xlim(-10, 10)
    plt.ylim(0, 1)
    plt.legend()
    plt.grid()
    plt.show()

#     
interact(
    plot_abs_normal_distribution,
    mean=FloatSlider(value=0, min=-10, max=10, step=0.1, description='Mean'),
    variance=FloatSlider(value=1, min=0.1, max=10, step=0.1, description='Variance')
)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# 1     
def plot_abs_distribution(mean, variance):
    std_dev = np.sqrt(variance)  #   

    # 1D  
    num_samples = 100000
    x = np.random.normal(mean, std_dev, num_samples)

    #  
    abs_x = np.abs(x)

    #  KDE 
    plt.figure(figsize=(8, 5))
    plt.hist(abs_x, bins=100, density=True, alpha=0.6, label='Sampled Data')

    # KDE 
    from scipy.stats import gaussian_kde
    kde = gaussian_kde(abs_x)
    abs_range = np.linspace(0, np.max(abs_x), 1000)
    plt.plot(abs_range, kde(abs_range), label='Estimated PDF', color='red', linewidth=2)

    plt.title(f'Absolute Value Distribution\nMean={mean:.2f}, Variance={variance:.2f}')
    plt.xlabel('|X|')
    plt.ylabel('Density')
    plt.xlim(-10, 10)
    plt.ylim(0, 1)
    plt.legend()
    plt.grid()
    plt.show()

#     
interact(
    plot_abs_distribution,
    mean=FloatSlider(value=0, min=-10, max=10, step=0.1, description='Mean'),
    variance=FloatSlider(value=1, min=0.1, max=10, step=0.1, description='Variance')
)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
from ipywidgets import interact, FloatSlider

# CVaR 
def compute_cvar(samples, alpha=0.9):
    sorted_samples = np.sort(samples)  #  
    VaR = np.percentile(sorted_samples, 100 * alpha)  # VaR 
    cvar = sorted_samples[sorted_samples >= VaR].mean()  # VaR   
    return VaR, cvar

#  CVaR 
def plot_abs_distribution_with_cvar(mean, variance, alpha=0.9):
    std_dev = np.sqrt(variance)  #   

    # 1D  
    num_samples = 100000
    x = np.random.normal(mean, std_dev, num_samples)
    abs_x = np.abs(x)  #  

    # CVaR 
    VaR, cvar = compute_cvar(abs_x, alpha)

    # KDE 
    kde = gaussian_kde(abs_x)
    abs_range = np.linspace(0, np.max(abs_x), 1000)
    pdf_values = kde(abs_range)

    # 
    plt.figure(figsize=(8, 5))
    plt.hist(abs_x, bins=100, density=True, alpha=0.6, label='Sampled Data')
    plt.plot(abs_range, pdf_values, label='Estimated PDF', color='red', linewidth=2)

    # VaR CVaR 
    plt.axvline(VaR, color='blue', linestyle='--', label=f'VaR ({int(alpha*100)}%) = {VaR:.2f}')
    plt.axvline(cvar, color='green', linestyle='--', label=f'CVaR ({int(alpha*100)}%) = {cvar:.2f}')
    plt.fill_between(abs_range, 0, pdf_values, where=(abs_range >= VaR), color='green', alpha=0.3, label='CVaR Region')

    plt.title(f'Absolute Value Distribution with CVaR\nMean={mean:.2f}, Variance={variance:.2f}')
    plt.xlabel('|X|')
    plt.ylabel('Density')
    plt.xlim(-10, 10)
    plt.ylim(0, 1)
    plt.legend()
    plt.grid()
    plt.show()

#     
interact(
    plot_abs_distribution_with_cvar,
    mean=FloatSlider(value=0, min=-10, max=10, step=0.1, description='Mean'),
    variance=FloatSlider(value=1, min=0.1, max=10, step=0.1, description='Variance')
)
```
