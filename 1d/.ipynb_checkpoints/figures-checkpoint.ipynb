```python
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np
from scipy import stats
from scipy import special
from tqdm import tqdm
```

# GT distribution of the obstacle

```python
mu = 5
sigma_2 = 2
sigma = np.sqrt(sigma_2)
gt_dist = norm(loc=mu, scale=sigma_2)

l = np.linspace(-20, 20, 100)
plt.plot(l, gt_dist.pdf(l))
plt.xlim(-25, 15)
plt.ylim(-0.1, 0.5)
plt.show()
```

# sa

```python
alpha = 0.95
l = np.linspace(-20, 20, 100)
plt.plot(l, gt_dist.pdf(l))

x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = stats.norm.pdf(x, mu, sigma_2)

VaR = gt_dist.ppf(alpha)
print(VaR)
new_VaR = np.sqrt(2) * special.erfinv(2*alpha-1)
print(mu + new_VaR * sigma)

beta = (np.sqrt(2*np.pi) * (np.exp(special.erfinv(2*alpha-1))**2) * (1-alpha)) ** (-1)
CVaR = mu + beta * sigma

plt.axvline(VaR, color='r', linestyle='--', label=f'VaR (α={alpha*100}%) = {VaR:.2f}')
plt.fill_between(x, 0, y, where=(x >= VaR), color='r', alpha=0.2)

plt.axvline(CVaR, color='g', linestyle='--', label=f'CVaR (α={alpha*100}%) = {CVaR:.2f}')
plt.fill_between(x, 0, y, where=(x >= CVaR), color='g', alpha=0.2)
plt.xlim(-25, 15)
plt.ylim(-0.1, 0.5)
plt.xlabel('Random Variable')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```
